### Carilah eigenvector dan eigenvalue dari dataset Anda. Rekomendasikan berapa jumlah principle component yang sebaiknya digunakan untuk dataset Anda, jika Anda menginginkan dimensi sekecil mungkin namun hasil plot kelas dapat terpisahkan dengan baik secara visual. Berikan analisis singkat mengapa Anda merekomendasikan hal tersebut.

#### data abalone_df terdiri dari 8 numerical atribut dan 1 nominal atribut yaitu 'sex'. setelah membaca dari sumber yang saya dapatkan,  ternyata PCA tidak dierkomendasikan untuk memproses categorical data. dataset yang ada atribut bertipe categorical lebih baik diproses menggunakan CATPCA. karena itu, saya akan menghilangkan atribut 'sex' dalam dataset saya selanjutnya untuk PCA.

sumber: https://www.researchgate.net/post/Should_I_use_PCA_with_categorical_data

```python
# drop sex category
abalone_num = abalone_df.copy()
abalone_num.drop('sex', axis=1, inplace=True)
abalone_num.head(5)
```

```python
# standarisasi data
from sklearn.preprocessing import StandardScaler

abalone_num = StandardScaler().fit_transform(abalone_num.iloc[:, :])
display(abalone_num)
```

```python
import numpy as np

# mendapatkan covariace matrix
abalone_cov_mat = np.cov(abalone_num.T)
display(abalone_cov_mat)
```

```python
# mendapatkan eigenvalue dan eigenvector

eig_vals, eig_vecs = np.linalg.eig(abalone_cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('Eigenvalues \n%s' %eig_vals)
```

```python
# make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range (len(eig_vals))]

# sort the (eigenvalue, eigenvector) tuples from high to low 
eig_pairs.sort()
eig_pairs.reverse()

# visualy confirm that the list is correctly sorted by decreasing values
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
# membuat explained variance
total_eig_vals = sum(np.abs(eig_vals))

print("Total sum of eigen values:")
display(total_eig_vals)
```

```python
var_exp = [(i / total_eig_vals) for i in sorted(np.abs(eig_vals), reverse=True)]

print("Explained variance:")
display(var_exp)
```

```python
cum_var_exp = np.cumsum(var_exp)

print("Cumulative sum of explained variance:")
display(cum_var_exp)
```

```python
# Visualisasi explained variance

plt.bar(range(1, eig_vals.size + 1), var_exp, alpha=0.5, align='center', label='Individual')
plt.step(range(1, eig_vals.size + 1), cum_var_exp, where='mid', label='Cumulative')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

#### pada visualisasi bar di atas, terlihat bahwa variance terbanyak ada pada PC yang pertama. jika dilihat pada cumulative ratio, line akan flat pada PC ke enam. dari hasil pengamatan tersebut, maka saya mengambil 5 PC pertama untuk membuat proyeksi PCA.

```python
var_exp_1 = var_exp[0] * 100
var_exp_2 = var_exp[1] * 100
var_exp_3 = var_exp[2] * 100
var_exp_4 = var_exp[3] * 100
var_exp_5 = var_exp[4] * 100
total_var_exp = var_exp_1 + var_exp_2 + var_exp_3 + var_exp_4 + var_exp_5

print(total_var_exp)
```

```python
from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(abalone_num)
```
